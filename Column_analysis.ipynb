```python
from string import punctuation
def address_corrector(address : str)->str:
    corrected_address = ""
    for char in address:
        if char not in punctuation:
            corrected_address+= char
        else:
            corrected_address+=" "
            
    corrected_address = corrected_address.upper()
    splited = corrected_address.split(" ")
    splited = [item for item in splited if item]
    return " ".join(splited)
```

```python
def create_season(month:int , day:int) -> str :
    if month in []:
        return "Winter"
    elif month in []:
        return "Spring"
    elif month in [] :
        return "Summer"
    else:
        return "Autumn"
```

```python
def seasons(m:int , d:int)-> str:
    season=None
    if (3==m and d>=21) or m==4 or m==5 or (m==6 and 20<=d):
        season = 'Spring'
    elif (6==m and d>=21 ) or m==7 or m==8 or (m==9 and 20<=d):
        season = 'Summer'
    elif (9==m and d>=21 ) or m==10 or m==11 or (m==12 and 20<=d):
        season = 'Autumn'
    elif (12==m and d>=21 ) or m==1 or m==2 or (m==3 and 20<=d):
        season = 'Winter'
    return season
```

```python
df_no_date["season"] = df_no_date.apply(lambda x: seasons(x.month , x.day) , axis=1)
```

```python
ddd = df_no_date.groupby(["year","season"] ,as_index=False).mean()
```
