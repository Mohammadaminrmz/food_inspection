```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from datetime import datetime, date, timedelta


dfm=pd.read_csv('Food_Establishment_Inspection_Data.csv')
#dfmprint(dfm.describe())
dfm[dfm.Business_ID== 'PR0088351'].head(5)
```

```python
dfins=dfm.groupby(by='Inspection_Serial_Num' , as_index=False).aggregate({'Business_ID': 'first','Inspection Score':'first', 'Inspection Date': 'first', })
dfins.rename(columns={'Inspection Date':'date' ,'Business_ID':'bid','Inspection Score':'score'}, inplace = True)
dfins['date']=pd.to_datetime(dfins['date'])
dfins.score=pd.to_numeric(dfins['score'])

dfins.head(3)
```

```python
df=dfm.loc[:, ['Business_ID','Inspection Date','Inspection Score','Longitude','Latitude','Inspection Closed Business']]
df.rename(columns={'Inspection Closed Business':'closed','Inspection Date':'date' ,'Business_ID':'bid','Inspection Score':'score'}, inplace = True)
df['date'] = pd.to_datetime(df['date'])

df.head(2)
# df.describe()
```

```python
df['long']=df['Longitude'].apply((lambda x: round(x,3)))
df['lat']=df['Latitude'].apply((lambda x: round(x,3)))
# df.drop(columns=['Latitude', 'Longitude'], inplace=True)
df.head(1)
```

```python
dfbid=df.drop_duplicates(subset='bid').dropna()
dfclosed=df[(df['closed']==True) &(df.date< '2021-01-09')]
dfbidclosed=dfclosed.drop_duplicates(subset='bid').dropna()
dfbid.info()
```

```python
print('all:' , len(df),'    all unique id:',len(dfbid),'   closed:',len(dfclosed),'  Closed unique id:', len(dfbidclosed))
```

```python

for i,index in zip(dfbidclosed.bid,dfbidclosed.index): #.iloc[0:2,]
    
    # closedate=dfbidclosed[dfbidclosed.bid==i].date.values[0]
    closedate=pd.to_datetime(dfbidclosed.loc[index,'date'])
    long=dfbidclosed.loc[index,'long']
    lat=dfbidclosed.loc[index,'lat']
    # print('----------------------',i,index,closedate)
    try:
        dffilterd=(dfbid[(dfbid.long== long) & (dfbid.lat==lat) ]).index #&  (dfbid.bid!=i)
      
        # print((dffilterd))
        if len(dffilterd)>1:
            dfbid.loc[dffilterd,'closedid']=i
            for indexid in dffilterd:
                bid=dfbid.loc[indexid,'bid']
                dfnearbefore=dfins[(dfins.bid==bid) &(dfins.date<closedate) &(dfins.date>(closedate- pd.offsets.DateOffset(years=2)))].aggregate({'bid': 'count', 'score': 'mean'})
                dfnearafter=dfins[(dfins.bid==bid) &(dfins.date>closedate) &(dfins.date<(closedate+ pd.offsets.DateOffset(years=2)))].aggregate({'bid': 'count', 'score': 'mean'})
                dfbid.loc[dffilterd,'closedid']=i
                # print(bid,indexid)
                
                # calculate formula:---------------------------------------------------------
                dfbid.loc[indexid,'countbefore']=dfnearbefore.bid
                dfbid.loc[indexid,'scorebefore']=round(dfnearbefore.score,0)
                dfbid.loc[indexid,'countafter']=dfnearafter.bid
                dfbid.loc[indexid,'scoreafter']=round(dfnearafter.score,0)
                dfbid.loc[indexid,'finalscore']=round(dfnearafter.score,0)+dfnearafter.bid*2-(round(dfnearbefore.score,0) + dfnearbefore.bid *2)
                # print((dfbid[dfbid.bid==bid]))
                # print(dfnearafter)

        
    except:
        print('error')
```

```python
df2=dfbid[(dfbid['closedid']!=0) &(dfbid['countbefore']>0)].sort_values('closedid').head(200) #200 restricts total places
# dfbid.groupby('closedid').count()
# dfbid[dfbid['closedid']=='PR0004034']
# df2.head(30)
```

```python
def colorfun(score,closeid,bid):
    if closeid == bid:
        return('red')
    else:
        if score >=0:
            return('blue')
        elif score<0:
            return('green')

df2['color']=df2.apply(lambda x: colorfun(x[13],x[8],x[0]) , axis=1)
df2.fillna(0, inplace=True)
df2.head(5)
```
