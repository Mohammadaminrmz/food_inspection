```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from datetime import datetime, date, timedelta


dfm=pd.read_csv('Food_Establishment_Inspection_Data.csv')
#dfmprint(dfm.describe())
dfm.head(2)
```

```python
df=dfm.loc[:, ['Business_ID','Inspection Date','City','Longitude','Latitude','Inspection Closed Business']]
df.rename(columns={'Inspection Closed Business':'closed','Inspection Date':'date' ,'Business_ID':'bid'}, inplace = True)
df.head(2)
# df.describe()
```

```python
df['long']=df['Longitude'].apply((lambda x: round(x,3)))
df['lat']=df['Latitude'].apply((lambda x: round(x,3)))
# df.drop(columns=['Latitude', 'Longitude'], inplace=True)
df.head(1)
```

```python
dfbid=df.drop_duplicates(subset='bid').dropna()
dfclosed=df[df['closed']==True]
dfbidclosed=dfclosed.drop_duplicates(subset='bid').dropna()

dfclosed.info()
```

```python
print('all:' , len(df),'    all unique id:',len(dfbid),'   closed:',len(dfclosed),'  Closed unique id:', len(dfbidclosed))
```

```python
dfbid['closedid']=0
```

```python

for i in dfbidclosed.bid:
    # print('----------------------',i,dfbid[dfbid.bid==i].long.values[0])
    
    # dfbid[]['closedid']=dfbid
    try:
        dffilterd=(dfbid[(dfbid.long== dfbid[dfbid.bid==i].long.values[0]) & (dfbid.lat==dfbid[dfbid.bid==i].lat.values[0]) ]).index #&  (dfbid.bid!=i)
        # print(len(dffilterd))
        # dfbid['closedid']
        # dfbid[dfbid.index.isin(dffilterd)].loc[:,'closedid'] = 4 #.apply(lambda x :i)
        if len(dffilterd)>1:
            
            dfbid.loc[dffilterd,'closedid']=i
        # print(dfff) #loc[:,'closedid'] = i
    except:
        print('error')
```

```python
df2=dfbid[dfbid['closedid']!=0].sort_values('closedid').head(30)
# dfbid.groupby('closedid').count()
# dfbid[dfbid['closedid']=='PR0004034']
df2.head(30)
```
